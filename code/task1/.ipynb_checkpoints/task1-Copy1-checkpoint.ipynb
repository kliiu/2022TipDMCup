```python
#不合并标题和正文
newdf.drop({"发布时间"},axis=1,inplace=True)
```

```python
#去除str中的空字符
strinfo = re.compile(' ')
#newdf["content"]=newdf["content"].apply(lambda x:strinfo.sub("",x))
newdf["公众号标题"]=newdf["公众号标题"].apply(lambda x:strinfo.sub("",x))
#newdf["正文"]=newdf["正文"].apply(lambda x:strinfo.sub("",x))
```

```python
#去除str中的数字
for row in range(621):
    #print("---------------"+str(row)+"--------------------")
    newdf["公众号标题"][row] = ''.join([i for i in newdf["公众号标题"][row] if not i.isdigit()])
    newdf["正文"][row] = ''.join([i for i in newdf["正文"][row] if not i.isdigit()])
    #print(newdf["content"][row])
```

```python
newdf
```

```python
#特殊符号
special_words=["\t","\n","★","↓","▼","…","|","◆","!","☆","▶","▷","█","●","👇"]
# 创建停用词列表
def stopwordslist():
    stopwords = [line.strip() for line in open('stopwords/hit_stopwords.txt',encoding='utf-8').readlines()]
    return stopwords
# 对句子进行中文分词
def seg_depart(sentence):
    # 对文档中的每一行进行中文分词
    sentence_depart = jieba.cut(str(sentence).strip())
    # 创建一个停用词列表
    stopwords = stopwordslist()
    # 输出结果为outstr
    outstr = ''
    # 去停用词
    for word in sentence_depart:
        if word not in stopwords:
            if word not in special_words:
                outstr += word
                outstr += " "
    return outstr
```

```python
#print(newdf["content"].iloc[0])
```

```python
for row in range(621):
    #print("---------------"+str(row)+"--------------------")
    newdf["公众号标题"][row]=seg_depart((newdf["公众号标题"].iloc[row]))
    newdf["正文"][row]=seg_depart((newdf["正文"].iloc[row]))
    #print(newdf["content"][row])
newdf
```

```python
#再将词典读出
new_dic=[]
with open('tempdic.txt','r')as file_obj:
    for line in file_obj.readlines():
        new_dic.append(line[:-1])#不读最后的换行符
```

```python
len(new_dic)
```

```python
#去除重复
lst = []
for el in new_dic:
    if lst.count(el) < 1:
        lst.append(el)
    else:
        print(el)
  # 没有问题,也能保留原有顺序,但是创建了新列表
```

```python
len(lst)
new_dic=lst
```

```python
#计算每篇文章标题各包含词表的单词个数列表
#相关的
dic_cnt=[]
ori_len=[]
for row in range(len(newdf)):
    d_cnt=0#计数
    o_cnt=0
    for word in (str(newdf.iloc[row,1]).split(" ")):
        o_cnt+=1
        if word in new_dic:
            d_cnt+=1
    dic_cnt.append(d_cnt)
    ori_len.append(o_cnt)
    
analysis_words=newdf.copy()
analysis_words["title_len"]=ori_len#原始词数
analysis_words["title_cnt"]=dic_cnt#出现词表中的词的次数
analysis_words["title_fre"]=analysis_words["title_cnt"]/analysis_words["title_len"]
analysis_words
```

```python
#计算每篇文章内容各包含词表的单词个数列表
#相关的
dic_cnt=[]
ori_len=[]
for row in range(len(newdf)):
    d_cnt=0#计数
    o_cnt=0
    for word in (str(newdf.iloc[row,2]).split(" ")):
        o_cnt+=1
        if word in new_dic:
            d_cnt+=1
    dic_cnt.append(d_cnt)
    ori_len.append(o_cnt)
    
analysis_words["content_len"]=ori_len#原始词数
analysis_words["content_cnt"]=dic_cnt#出现词表中的词的次数
analysis_words["content_fre"]=analysis_words["content_cnt"]/analysis_words["content_len"]
analysis_words
```

```python
#计算最终概率
analysis_words["fre"]=analysis_words["content_fre"]*0.4+analysis_words["title_fre"]*0.6
analysis_words
```
